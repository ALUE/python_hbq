### （1）导入matplotlib

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

## %matplotlib inline

### （2）结合pandas画图

```python
pd.value_counts(table[u"本科专业"])
```

```python
def set_ch():
	from pylab import mpl
	mpl.rcParams['font.sans-serif'] = ['SimHei'] # 指定默认字体
	mpl.rcParams['axes.unicode_minus'] = False # 解决保存图像是负号'-'显示为方块的问题
set_ch()
```

```python
#-*- coding: utf-8 -*-
tab1 = pd.value_counts(table[u"本科专业"])
tab2 = tab1[tab1 > 2]

fig, ax = plt.subplots()
y_pos = np.arange(len(tab2))

ax.barh(y_pos, tab2.values)

ax.set_yticks(y_pos)
ax.set_yticklabels(tab2.index)

plt.show()
```

```python
tab2 = tab1[tab1 < 2]

fig, ax = plt.subplots()
y_pos = np.arange(len(tab2))

ax.barh(y_pos, tab2.values)

ax.set_yticks(y_pos)
ax.set_yticklabels(tab2.index)

plt.show()
```

```python
computer_majors = [u'软件工程',u'计算机科学与技术',u'网络与信息安 ',u'物联网工程',u'信息安全']

def getMajor(x):
    return x in computer_majors
table[u'计算机相关专业'] = table[u'本科专业'].apply(lambda x: getMajor(x))
#table
```

```python
labels=['Compute','Non-Computer']  
X=[sum(table[u'计算机相关专业'] > 0),sum(table[u'计算机相关专业'] == 0)]    
  
fig = plt.figure()  
plt.pie(X,labels=labels)##,autopct='%1.2f%%') #画饼图（数据，数据对应的标签，百分数保留两位小数点）  

plt.show()  
```

## K-means

```python
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

plt.figure(figsize=(12, 12))

n_samples = 1500
random_state = 170
X, y = make_blobs(n_samples=n_samples, random_state=random_state)
```

```python
plt.figure(figsize=(10, 8))

plt.scatter(X[:, 0], X[:, 1])
```

```python
y_pred = KMeans(n_clusters=3).fit_predict(X)
```

```python
plt.figure(figsize=(10, 8))
rgb = ['r','g','b']

for i in range(3):
    X2 = np.zeros((500,2))
    k = 0
    for j in range(1500):
        if y_pred[j] == i:
            X2[k,0] = X[j, 0]
            X2[k,1] = X[j, 1]
            k+=1
    plt.scatter(X2[:, 0], X2[:, 1], c = rgb[i])
```
